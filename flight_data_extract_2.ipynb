```python
url='https://www.makemytrip.com/flight/search?tripType=O&itinerary=AMD-DEL-27/10/2021&paxType=A-1_C-0_I-0&cabinClass=E&sTime=1635268619363&forwardFlowRequired=true&mpo=&intl=false'
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome("chromedriver.exe")
driver.implicitly_wait(20)
driver.get(url)
soup=BeautifulSoup(driver.page_source, 'lxml')
##First grabbing the url to access the web, using the webdriver

name1=soup.find_all('span',class_='boldFont blackText airlineName') ### getting the names of the Airlines
Airlines1=[]
for i in name1:
    Airlines1.append(i.get_text())
    
print(len(Airlines1))  

dep_time1=soup.find_all('p',class_='blackText fontSize18 blackFont appendBottom2 makeFlex hrtlCenter') ### getting the departure and arrival times of the Airlines
dep1, arr1 =dep_time1[::2], dep_time1[1::2]##Splitting the list into two as it was getting both the imformation together
DEP1=[]
for i in dep1:
    DEP1.append(i.get_text())
    ARV1=[]
for i in arr1:
    ARV1.append( i.get_text())
print(len(DEP1),len(ARV1))

layover1=soup.find_all('p',class_='flightsLayoverInfo')### getting the layover taken by the Airlines.
Lay_over1=[]
for l in layover1:
    Lay_over1.append(l.get_text())

dep_place1=soup.find_all('p', class_='darkText')### getting the departure and arrival plces of the Airlines
dep_place1, arr_place1 =dep_place1[::2], dep_place1[1::2]##Splitting the list into two as it was getting both the imformation together
origin1=[]
for i in dep_place1:
    origin1.append(i.get_text())
    destination1=[]
for i in arr_place1:
    destination1.append(i.get_text())
print(len(origin1),len(destination1))

price1=[]
pri1=soup.find_all('p',class_='blackText fontSize18 blackFont white-space-no-wrap')### getting the price of the Airlines
for i in pri1:
    price1.append(i.get_text())
    print(len(price1))


try:    
    dur1=driver.find_elements_by_xpath('//div[@class="stop-info flexOne"]/p')### getting the duration of the Airlines
    Dur1=[]
    for i in dur1:
        Dur1.append(i.text)
        print(len(Dur1))
except StaleElementReferenceException as e:
    print("Exception raised", e)
```

##### creating the dataset.
-- later these will be concatinated together.

```python
df1=pd.DataFrame()
df1['Airline_name']= Airlines1[0:23]
df1['Departure']=DEP1[0:23]
df1['Arrival']=ARV1[0:23]
df1['Source']=origin1[0:23]
df1['Destination']=destination1[0:23]
df1['Lay_over']=Lay_over1[0:23]
df1['Price']=price1[0:23]
df1['Duration']=Dur1[0:23]
df1
```
